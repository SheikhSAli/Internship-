1. Scrape the details of most viewed videos on YouTube from Wikipedia:<br>
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos<br>
You need to find following details:<br>
A) Rank<br>
B) Name<br>
C) Artist<br>
D) Upload date<br>
E) Views<br>

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Titles = []
Artists = []
Views = []
Upload_date = []
Ranks = []
for i in range(2,32):
    
    time.sleep(3)
    
    try:
        rank = driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[1]').text
        Ranks.append(rank)
    except:
        Ranks.append('NA')
    try:
        title = driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[2]').text
        Titles.append(title)
    except:
        Titles.append('NA')
    try:
        artist = driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[3]').text
        Artists.append(artist)
    except:
        Artists.append('NA')
    try:
        view = driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[4]').text
        Views.append(view)
    except:
        Views.append('NA')
    try:
        date = driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[5]').text
        Upload_date.append(date)
    except:
        Upload_date.append('NA')

youtube_Wiki_df = pd.DataFrame(list(zip(Ranks , Titles , Artists , Views , Upload_date )),
                               columns=['Ranks' , 'Titles' , 'Artists' , 'Views' , 'Upload_Date'  ])
youtube_Wiki_df
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.naukri.com/')
```

```python
recruiter = driver.find_element_by_xpath('//*[@id="root"]/div[1]/div/ul[1]/li[2]/a').get_attribute('href')
driver.get(recruiter)
```

```python
search_submit = driver.find_element_by_xpath('//*[@id="qsbFormBtn"]')
search_submit.click()
```

```python
soup = BeautifulSoup(driver.page_source , 'lxml')
all_d = soup.find_all('div' , class_ = 'outerRecSec')
```

```python
Names = []
Designation = []
location = []
Skills = []
for j in all_d:
    Name = j.find('span' , class_ = 'fl ellipsis')
    Names.append(Name.text)
    desig = j.find('span' , class_ = 'ellipsis clr')
    Designation.append(desig.text)
    loc = j.find('small' , class_ = 'ellipsis')
    location.append(loc.text)
    skill = j.find('div' , class_ = 'hireSec highlightable')
    Skills.append(skill.text)
# print(Names)
# print(Designation)
# print(location)
# print(Skills)
Naukri_df = pd.DataFrame({'R_Names':Names , 'R_Designation' : Designation , 'R_Location' :location , 'R_Skills':Skills})
Naukri_df
```
