<font color = 'Orange'>**Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:<br>1. Brand<br>
2. Product Description<br>
3. Price<br>
4. Discount %**</font>


```python
driver = webdriver.Chrome("chromedriver.exe")
url = 'https://www.flipkart.com'
driver.get(url)
driver.set_page_load_timeout(4)
```

```python
search_inp = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_inp.send_keys('Sunglasses')
```

```python
search_click = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_click.click()                                            
```

```python
soup = BeautifulSoup(driver.page_source , 'lxml')
```

```python
all_d = soup.findAll('div' , attrs = {'class' : '_13oc-S _1t9ceu'})
all_d
```

```python
B_df = []
counter =  0
while counter<3 :
    all_d = soup.findAll('div' , attrs = {'class' : '_1xHGtK _373qXS'})
    for d in all_d:
        try:
            name = d.find( 'div', attrs = {'class' : '_2WkVRV'}).text
            desc = d.find('a' , attrs = {'class' : 'IRpwTa'}).text.strip()
            disc_price = d.find('div' , attrs = {'class' : '_30jeq3'}).text
            orig_price = d.find('div' , attrs = {'class' : '_3I9_wc'}).text
            B_df.append({'Brand' : name , 'Short_desc' : desc , 'Discounted_price': disc_price , 'Original_price' : orig_price})
        except:
            pass
    try:
        next_page = driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
        next_page.click()
        page = requests.get(next_page)
        soup = BeautifulSoup(page.page_source , 'lxml')
    except:
        pass
    counter+=1

Brand_df = pd.DataFrame(B_df) 
Brand_df
```
